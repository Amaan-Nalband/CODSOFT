```python
#split the data set on differnt flowers
df_setosa=df.loc[df["species"]=="Iris-setosa"]
df_versicolor=df.loc[df["species"]=="Iris-versicolor"]
df_virginica=df.loc[df["species"]=="Iris-virginica"]
#compute PDF,CDF on petal length for setosa flower
counts,bin_edge=np.histogram(df_setosa['petal_length'],bins=10,density=True)
# To Calculate P.D.F
pdf=counts/sum(counts)
print("PDF on petal length for setosa flower",pdf)
cdf=np.cumsum(pdf)
print("CDF on petal length for setosa flower",cdf)
#plot the pdf and cdf for setosa
plt.xlabel("Pdf of Setosa")
plt.ylabel("Cdf of Setosa")
plt.plot(bin_edge[1:],pdf,label='PDF')
plt.plot(bin_edge[1:],cdf,label='CDF')
plt.show()
```

```python
#statitical computation
# mean (tell us the central tendency of data distribution)
mean_of_setosa=np.mean(df_setosa["petal_length"])
# Similar we compute for verginica and versicolor
mean_of_virginica=np.mean(df_virginica["petal_length"])
mean_of_versicolor=np.mean(df_versicolor["petal_length"])
print("Means:\n")
print(mean_of_setosa)
print(mean_of_virginica)
print(mean_of_versicolor)

print("\nStd-dev:")
print(np.std(df_setosa["petal_length"]));
print(np.std(df_virginica["petal_length"]))
print(np.std(df_versicolor["petal_length"]))

print("\nMedians:")
print(np.median(df_setosa["petal_length"]))
#mean with an outlier
print(np.median(np.append(df_setosa["petal_length"],50)));
print(np.median(df_virginica["petal_length"]))
print(np.median(df_versicolor["petal_length"]))
```

```python
print("\nQuantiles:")
print(np.percentile(df_setosa["petal_length"],np.arange(0, 100, 25)));
print(np.percentile(df_virginica["petal_length"],np.arange(0, 100, 25)))
print(np.percentile(df_versicolor["petal_length"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(df_setosa["petal_length"],90));
print(np.percentile(df_virginica["petal_length"],90))
print(np.percentile(df_versicolor["petal_length"],90))

from statsmodels import robust
print("\nMedian Absolute Deviation")
print(robust.mad(df_setosa["petal_length"]))
print(robust.mad(df_virginica["petal_length"]))
print(robust.mad(df_versicolor["petal_length"]))
```
